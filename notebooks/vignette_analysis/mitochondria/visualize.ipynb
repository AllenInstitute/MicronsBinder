# Mitochondria visualization
A brief notebook to demonstrate visualizing a cell id and its mitochondria

```python
import pandas as pd
from nglui import statebuilder as sb

from lib import ngl
```

Reading the dataframe that links mitochondria to cell IDs

```python
mitodf = pd.read_csv("data/pni_mito_analysisids_fullstats.csv")
```

```python
mitodf.head()
```

```python
img = sb.ImageLayerConfig(name="image", source=ngl.IMG_SOURCE)
seg = sb.SegmentationLayerConfig(name="seg", source=ngl.SEG_SOURCE,
                                 selected_ids_column="cellid",
                                 view_kws=dict(alpha_3d=0.25))  # some default transparency
mito = sb.SegmentationLayerConfig(name="mitochondria", source=ngl.MITO_SOURCE,
                                  selected_ids_column="mito_id")
```

```python
builder = sb.StateBuilder(layers=[img, seg, mito],
                          view_kws=dict(layout="3d", zoom_3d=1000))
```

```python
builder.render_state(mitodf[mitodf.cellid == 648518346349539840], return_as="html")
```
